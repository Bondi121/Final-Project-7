```python
import pandas as pd
```

```python
data_2010 = pd.read_csv("2010.csv")
data_2011 = pd.read_csv("2011.csv")
data_2012 = pd.read_csv("2012.csv")
data_2013 = pd.read_csv("2013.csv")
data_2014 = pd.read_csv("2014.csv")
data_2015 = pd.read_csv("2015.csv")
data_2016 = pd.read_csv("2016.csv")
data_2017 = pd.read_csv("2017.csv")
data_2018 = pd.read_csv("2018.csv")
data_2019 = pd.read_csv("2019.csv")
```

### Clean the dataset to make sure the following columns are either integer or floats and not objects:


```python
data_2010.info()
```

```python
all_data = pd.concat([data_2010, data_2011, data_2012, data_2013, data_2014, data_2015, data_2016, data_2017, data_2018, data_2019])
all_data.head(3)
```

```python
all_data.info()
```

```python
all_data.shape
```

```python
all_data.rename(columns = {'Unnamed: 0':'ID'}, inplace = True)
```

```python
all_data["title"] = all_data["title"].str.strip()
all_data["artist"] = all_data["artist"].str.strip()
all_data["top genre"] = all_data["top genre"].str.strip()
```

```python
pd.set_option("display.max_rows", None)
```

```python
all_data
```

Column names:

ID = unique id number

title = song’s title

artist = song’s artist

top genre = the genre of the track

year = song’s year in the billboard

bpm = beats per minute (tempo)

nrgy = energy, the higher the value the more energetic the song

dnce = danceability, the higher the value the easier it is to dance to

db = loudness, the higher the value the louder the song

live = the higher the value the more likely the song is a live recording

val = valence, the higher the value the more positive the mood

dur = duration of the song

acous = acousticness, the higher the value the more acoustic the song is

spch = speechiness, the higher the value the more spoken word the song contains

pop = popularity, the higher the value the more popular the song is

The new dataframe all_data is merged from various csv files, using the pd concat method. Whitespaces where stripped for the three object columns and the other columns are integers as they are supposed to. By using df.info and looking through the data set, the le looks properly cleaned.

### What percentage of songs may have been recorded live? (Consider anything above 50 in the live column to be recorded live)

```python
column = all_data["live"]
count = column[column > 50].count()
count
```

```python
count_neg = column[column < 50].count()
count_neg
```

```python
(100 / count_neg * count).round(2)
```

The percentage of songs recorder live is 3.08%. Using the count functions for live songs above 50 in that column and under 50 helped to find the absolute number for each value. After that finding the percentage was fairly straight forward.

### What percentage of songs were acoustic? (Consider anything above 50 in the acoustic column to be acoustic)

```python
column = all_data["acous"]
count_acous = column[column > 50].count()
count_acous
```

```python
count_neg_acous = column[column <= 50].count()
count_neg_acous
```

```python
(100 / count_neg_acous * count_acous).round(2)
```

7.52 % of songs were acoustic. The same method was applied in the previous answer.

### Which year’s songs have the highest amount of spoken word in the song?

```python
highest_amount_spoken = all_data[['year', 'spch']].groupby('year').sum()
highest_amount_spoken  = highest_amount_spoken.sort_values('spch', ascending=False)
highest_amount_spoken 
```

2015 has the highest amount of spoken word in the song with 670. Grouping the year and displaying the highest amount of spoken words in total with the sum function was applied.
